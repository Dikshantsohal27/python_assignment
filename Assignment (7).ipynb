```python
# i select 10 column for clear visualization
# in this we try to find the rating of the code 162  
filtered_data = df[df['Country Code'] == 162]
filtered_data.head(10)

plt.hist(filtered_data['Aggregate rating'], bins=10, edgecolor='black')
plt.title('Distribution of Restaurant Ratings in Country Code 162')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# we drop it because 94 code is the outlier
df = df[df['Country Code'] != 94]
```

```python
# avgerage order price from each country to see how wealthy the country is
avg_order = df.groupby('Country Code')['Average Cost for two'].mean()

avg_order.plot(kind='bar', edgecolor='black')
plt.title('Average Order Amounts by Country')
plt.xlabel('Country Code')
plt.ylabel('Average Order Amount')
plt.show()
```

```python
# rating of the resturant in difff category
x = df['Rating text'].value_counts()
x.plot(kind='bar',edgecolor='black')
plt.title('Summary')
plt.xlabel('Rating')
plt.ylabel('No of resturant')
plt.show()
```
